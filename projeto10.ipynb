# Passo 1: Importando Bibliotecas

```python
# Importando bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Passo 2: Carregando os Dados

```python
# Carregando os dados
url = 'https://raw.githubusercontent.com/andre-marcos-perez/ebac-course-utils/develop/dataset/credito.csv'
df = pd.read_csv(url, na_values='na')
```

****Explicação:**** Estamos carregando os dados do conjunto de dados fornecido. Certifique-se de substituir o link pelo link do seu conjunto de dados, se estiver usando outro.

# Passo 3: Função de Limpeza de Dados

```python
# Função para limpeza de dados
def limpar_dados(df):
    # Correção de tipo de dados para colunas específicas
    df['limite_credito'] = df['limite_credito'].apply(lambda x: float(x.replace(".", "").replace(",", ".")))
    df['valor_transacoes_12m'] = df['valor_transacoes_12m'].apply(lambda x: float(x.replace(".", "").replace(",", ".")))

    # Remoção de linhas com dados faltantes
    df.dropna(inplace=True)
```

```python
# Aplicando a função de limpeza
limpar_dados(df)
```

****Explicação:**** Esta função realiza a limpeza de dados, corrigindo o tipo de dados e removendo linhas com dados faltantes.

# Passo 5: Mostrar estatísticas resumidas após a limpeza

```python
print("Estatísticas resumidas após a limpeza:")
print(df.describe())

# Verificar dados faltantes após a limpeza
print("\nDados faltantes após a limpeza:")
print(df.isnull().sum())
```

****Explicação:**** essas operações fornecem uma visão geral das estatísticas descritivas das colunas numéricas do DataFrame após a limpeza e mostram a quantidade de dados faltantes em cada coluna, o que é útil para avaliar a qualidade dos dados e identificar áreas que podem precisar de mais atenção ou imputação.

# Passo 8: Visualizações Categóricas

```python
# Visualizações categóricas
sns.set(style="whitegrid")
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# Distribuição por Escolaridade
sns.countplot(x='escolaridade', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Distribuição por Escolaridade')

# Distribuição por Estado Civil
sns.countplot(x='estado_civil', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Distribuição por Estado Civil')

# Distribuição de Tipo de Cartão
sns.countplot(x='tipo_cartao', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Distribuição por Tipo de Cartão')

# Distribuição de Dependentes
sns.countplot(x='dependentes', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Distribuição por Dependentes')

plt.tight_layout()
plt.show()
```

****Explicação:**** Este bloco cria visualizações categóricas para entender a distribuição de diferentes variáveis categóricas.

# Passo 9: Visualizações Numéricas

```python
# Visualizações numéricas
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# Distribuição de Idade
sns.histplot(x='idade', data=df, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribuição de Idade')

# Distribuição de Salário Anual
sns.histplot(x='salario_anual', data=df, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribuição de Salário Anual')

# Quantidade de Transações nos Últimos 12 Meses
sns.histplot(x='qtd_transacoes_12m', data=df, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Quantidade de Transações nos Últimos 12 Meses')

# Valor das Transações nos Últimos 12 Meses
sns.histplot(x='valor_transacoes_12m', data=df, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Valor das Transações nos Últimos 12 Meses')

plt.tight_layout()
plt.show()
```

****Explicação:**** Este bloco cria visualizações numéricas para entender a distribuição de diferentes variáveis numéricas.
